# Exploratory Data Analysis
First up, when importing a netcdf file, it is important to understand what the coordinate systems, what variables are included, and what the bounds are.

```python
import xarray as xr
import numpy as np

file = '../data/sos_Omon_HadCM3_past1000_r1i1p1_085001-185012.nc'
#file = "https://esgf.ceda.ac.uk/thredds/fileServer/esg_dataroot/pmip3/output/UOED/HadCM3/past1000/mon/ocean/Omon/r1i1p1/v20130313/sos/sos_Omon_HadCM3_past1000_r1i1p1_085001-185012.nc#mode=bytes"
ds = xr.load_dataset(file)
ds
```

As we can see, the data uses Latitude and Longitude at 1.25 degree intervals, and time every month from January 850 to December 1850. Sea surface salinity, sos, is the variable of import. 

Next, we want to look at the extent of coverage and compare that to the locations we are using - the cores.txt file.

```python
import pandas as pd
cores = pd.read_table('../data/cores.txt')
cores
```

```python
import cartopy as cr # for geographic mapping
import cartopy.crs as ccrs # for map projections
import matplotlib.pyplot as plt # plotting tool
import cartopy.feature as cfeature # to add coastlines, land and ocean
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

sos2d = ds.sos[12011,:,:]

fig = plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())
ax.set_extent([-180,180,-90,90],ccrs.PlateCarree())
ax.add_feature(cfeature.COASTLINE,lw=.5)
ax.set_xticks([-180,-120,-60, 0, 60, 120, 180], crs=ccrs.PlateCarree())
ax.set_yticks([-90, -60, -30, 0, 30, 60, 90], crs=ccrs.PlateCarree())
cmap=plt.cm.viridis_r
clev = np.arange(30,40,1)
pc=ax.contourf(ds.lon,ds.lat,sos2d,clev,transform=ccrs.PlateCarree(),cmap=cmap,extend='both')
ttitle='Salinity for 1850-12'
ax.set_title(ttitle,loc='left',fontsize=9)
ax.scatter(cores.Long,cores.Lat,color="red")
cax = plt.axes([0.92, 0.29, 0.04, 0.4])
plt.colorbar(pc,cax=cax)
plt.show()
```

The coverage is global, but does not include landmasses, and at the scale has incongruities with the coastline. While several of the core sites are located near to the coast, they do not seem to be out of bounds of data found in this model.

Next, testing the selection of data at given locations to see what the output is.

```python
ds.sos.sel(lon=0, lat=0, method='nearest')
```

```python
sss = ds.sos.sel(lat=10.5, lon=57.5, method='nearest')
sss.max('time').values.ravel()
```

Given a set longitude and latitude, the 'nearest' finds the closest value, as the scale of core sites does not match up with the scale of the netcdf file. In this case, it is off by 0.625 degrees latitude. The output is an array of sea surface salinity values for all times. This array can be used for various functions, such as the .max() and .min(), but require .values.ravel() to get directly instead of including the metadata.

Longitude values in cores.txt are -180 to 180, whereas the netcdf files are in 0 to 360, so these need to be transformed.

```python
for x in range(cores.shape[0]):
    lon = cores.Long[x]
    if lon<0:
        lon = lon+360
        cores.loc[x,'Long'] = lon 
cores
```

With the longitude transformed, the summary statistics can be calculated into empty columns.

```python
cores['sss_max'] = np.nan  
cores['sss_min'] = np.nan
cores['sss_avg'] = np.nan
cores['sss_std'] = np.nan  
```

```python
for x in range(cores.shape[0]):
    lat = cores.Lat[x]
    lon = cores.Long[x]
    sss = ds.sos.sel(lon=lon, lat=lat, method='nearest')
    cores.loc[x,'sss_max'] = sss.max('time').values
    cores.loc[x,'sss_min'] = sss.min('time').values
    cores.loc[x,'sss_avg'] = np.average(sss.values)
    cores.loc[x,'sss_std'] = np.std(sss.values)
cores
```

We can see all 58 cores, and use this information for further statistical analysis. For example, identifying which cores have high standard deviations.

```python
for x in range(cores.shape[0]):
    if cores.sss_std[x] > 1:
        print(cores.Core[x])
```
