In addition to the files downloaded directly to the data folder, it is possible to pull directly via the download url. However, the netcdf files all require "#mode=bytes" to be added at the end in order to be read. This has been included directly in to where they are read by the **combo** function. 

The GISS download links all have a standardized url, with the variable in the experiment number (121-128) and the time frames (850-1800).

```python
urldict = {}

for i in range (121,129):
    urldict['GISS_'+str(i)]=[]
    file_base = "https://ds.nccs.nasa.gov/thredds/fileServer/CMIP5/NASA/GISS/past1000/E2-R_past1000_r1i1p" + str(i) + "/sos_Omon_GISS-E2-R_past1000_r1i1p"
    for j in range(850,1000,50):
        urldict['GISS_'+str(i)].append(file_base + str(i) + "_0" + str(j) + "01-0" + str(j+49) + "12.nc#mode=bytes")
    for j in range(1000,1050,50):
        urldict['GISS_'+str(i)].append(file_base + str(i) + '_' + str(j) + '01-' + str(j+50)+'12.nc#mode=bytes')
    for j in range(1051,1851,50):
        urldict['GISS_'+str(i)].append(file_base + str(i) + '_' + str(j) + '01-' + str(j+49)+'12.nc#mode=bytes')
```
